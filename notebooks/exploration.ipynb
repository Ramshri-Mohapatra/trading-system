```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
os.chdir("C:/Projects/trading-system")

df = pd.read_csv("data/raw/AAPL.csv", header=[0,1], index_col=0, parse_dates=True)
print( "Shape", df.shape)

df.head()
```

```python
print("\nColumn Names: ")
print(df.columns)

print("\nData Types: ")
print(df.dtypes)

df['Close', 'AAPL'].plot(title="AAPL Close Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price(USD)")
plt.show()
```

```python


df_aapl = pd.read_csv("data/raw/AAPL.csv", header = [0,1], index_col=0, parse_dates=True)
df_msft = pd.read_csv("data/raw/MSFT.csv", header = [0,1], index_col=0, parse_dates=True)
df_tsla = pd.read_csv("data/raw/TSLA.csv", header = [0,1], index_col=0, parse_dates=True)

ax = df_aapl['Close','AAPL'].plot(label = 'AAPL', title='Stock Prices 2020-2024')
df_msft['Close', 'MSFT'].plot(label = 'MSFT')
df_tsla['Close', 'TSLA'].plot(label='TSLA')
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.legend()
plt.show()
```

```python
#Highest price AAPL ever reached in the period
highest = df[('Close','AAPL')].max()
print(f"Highest stock price of apple over the period: {highest}")

#Worst single day return
daily_returns = df[('Close', 'AAPL')].pct_change()
worst_day = daily_returns.min()
print(f"Worst single day drop: {worst_day:.2%} ")

#highest average close price over the 4 years.
aapl_avg = df_aapl[('Close', 'AAPL')].mean()
msft_avg = df_msft[('Close', 'MSFT')].mean()
tsla_avg = df_tsla[('Close', 'TSLA')].mean()

print(f"AAPL average: {aapl_avg:.2f}")
print(f"MSFT average: {msft_avg:.2f}")
print(f"TSLA average: {tsla_avg:.2f}")

#How many trading days are in the dataset
print(f"Total trading days {len(df)}")

#Best single day gain for apple
best_day = daily_returns.max()
print(f"Best single day return for apple: {best_day:.2%}")

# Apples's volatility
volatility = daily_returns.std()
print(f"APPL's volatility: {volatility:.2%}")

#Annual volatility of apple stocks
annual_volatility = volatility * (252 ** 0.5)
print(f"AAPL annual volatility: {annual_volatility:.2%}")





```
